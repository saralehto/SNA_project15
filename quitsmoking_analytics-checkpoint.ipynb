```python
text = data['text']
mentions = data.text.str.findall(r'@([A-Za-z0-9_]+)') #(?<=^|(?<=[^a-zA-Z0-9-\.]))

mentions_list = []
empty = []
for i in mentions.items():
    if i[1] != empty:
        mentions_list.append(i[1])

users = []
for mentions in mentions_list:
    for user in mentions:
        if user not in users:
            users.append(user)
nodes = users
```

```python
import networkx as nx
G = nx.DiGraph()
G.add_nodes_from(users)

from itertools import combinations
edges = []
while mentions_list != []:
    x = 0
    mention = []
    mention = mentions_list.pop()
    for k in mention:
        x += 1
        if x != 1:
            pairs = list(combinations(mention, 2))
            while pairs != []:
                pair = pairs.pop()
                edges.append(pair)
                
G.add_edges_from(edges)

pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=5, node_color='b')
nx.draw_networkx_edges(G, pos, alpha=0.4)
plt.show()
```

```python
no_nodes = G.number_of_nodes()
print(no_nodes)
no_edges = G.number_of_edges()
print(no_edges)
```
