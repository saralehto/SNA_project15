```python
hashtags = data.text.str.findall(r'#.*?(?=\s|$)')
print(hashtags)

tags = [] # tags is a list of every tweets' hashtags in an order
for t in hashtags:
    if t not in tags:
        tags.append(t)

hashs = [] # hashs is a list of different hashtags in the tweets
while tags != []:
    tag = tags.pop()
    for t  in tag:
        if t not in hashs:
            hashs.append(t)
nodes_tags = hashs
```

```python
G_tags = nx.DiGraph()
G_tags.add_nodes_from(hashs)

hashtags_same_tweet = []
while hashs != []:
    x = 0
    hashs_item = []
    hashs_item = hashs.pop()
    for h in hashs_item:
        x += 1
        if x != 1:
            pairs = list(combinations(hashs_item, 2))
            while pairs != []:
                pair = pairs.pop()
                hashtags_same_tweet.append(pair)

G_tags.add_edges_from(hashtags_same_tweet)

pos = nx.spring_layout(G_tags)
nx.draw_networkx_nodes(G_tags, pos, node_size=5, node_color='b')
nx.draw_networkx_edges(G_tags, pos, alpha=0.4)
plt.show()
```

```python
no_nodes_tags = G_tags.number_of_nodes()
print(no_nodes_tags)
no_edges_tags = G_tags.number_of_edges()
print(no_edges_tags)
```
